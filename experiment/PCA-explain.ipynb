```python
import pandas as pd

df = pd.read_csv('dataset/PUE.csv')
df
```

```python
df.drop(columns=["SSID"], inplace=True)
# df
```

```python
X = df.drop(columns=["logEB", "logYM", "logTS"])
X
```

```python
from hyperimpute.plugins.utils.simulate import simulate_nan
import numpy as np

np.random.seed(0)

X_with_nan = simulate_nan(X.values, 0.1, mecha="MAR", p_obs=0.0)['X_incomp']
X_with_nan = pd.DataFrame(X_with_nan, columns=X.columns)
```

```python
# find the index of the row with the nan values
nan_index = X_with_nan.isnull().any(axis=1)
nan_index
```

```python
from hyperimpute.plugins.imputers import Imputers, ImputerPlugin
from Impute import fill_with_et

imputers = Imputers()


class EtImputer(ImputerPlugin):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self._model = fill_with_et

    @staticmethod
    def name():
        return "et"

    @staticmethod
    def hyperparameter_space():
        return []

    def _fit(self, *args, **kwargs):
        return self

    def _transform(self, df):
        # 按照缺失值的比例进行排序
        miss_rate = df.isnull().sum() / df.shape[0]
        cols = miss_rate.sort_values().index.tolist()
        cols = [col for col in cols if miss_rate[col] > 0]
        for col in cols:
            df_col_filled = self._model(df, col)
            df[col] = df_col_filled[col]
        return df


imputers.add("et", EtImputer)

et = imputers.get("et")
missforest = imputers.get("missforest")
hyperimpute = imputers.get("hyperimpute")
mean = imputers.get("mean")
gain = imputers.get("gain")
sinkhorn = imputers.get("sinkhorn")

X_filled_et = et.fit_transform(X_with_nan.copy())
X_filled_missforest = missforest.fit_transform(X_with_nan.copy())
X_filled_hyperimpute = hyperimpute.fit_transform(X_with_nan.copy())
X_filled_gain = gain.fit_transform(X_with_nan.copy())
X_filled_sinkhorn = sinkhorn.fit_transform(X_with_nan.copy())
X_filled_mean = mean.fit_transform(X_with_nan.copy())
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'no-latex'])

plt.figure(figsize=(10, 8))

pca = PCA(n_components=2)
pca_func = pca.fit(X)
pca_1_ratio = pca.explained_variance_ratio_[0]
pca_2_ratio = pca.explained_variance_ratio_[1]

X_pca = pca_func.transform(X)

# plot the nan rows tsne
X_pca_nan = X_pca[nan_index]

X_filled_et_pca = pca_func.transform(X_filled_et)
X_filled_missforest_pca = pca_func.transform(X_filled_missforest)
X_filled_hyperimpute_pca = pca_func.transform(X_filled_hyperimpute)
X_filled_gain_pca = pca_func.transform(X_filled_gain)
X_filled_sinkhorn_pca = pca_func.transform(X_filled_sinkhorn)
X_filled_mean_pca = pca_func.transform(X_filled_mean)

X_filled_et_pca_nan = X_filled_et_pca[nan_index]
X_filled_missforest_pca_nan = X_filled_missforest_pca[nan_index]
X_filled_hyperimpute_pca_nan = X_filled_hyperimpute_pca[nan_index]
X_filled_gain_pca_nan = X_filled_gain_pca[nan_index]
X_filled_sinkhorn_pca_nan = X_filled_sinkhorn_pca[nan_index]
X_filled_mean_pca_nan = X_filled_mean_pca[nan_index]

# ["#F35F5F", "#7ED9D9", "#9467BD", "#B3DE69", "#FFC0D9", "#5FBDFF", "#FDBF6F"]

# plot the nan rows tsne

sns.scatterplot(x=X_pca_nan[:, 0], y=X_pca_nan[:, 1], s=120, marker='o', linewidths=2, facecolors='#7ED9D9',
                edgecolors='#7ED9D9',label="Original")

sns.scatterplot(x=X_filled_hyperimpute_pca_nan[:, 0], y=X_filled_hyperimpute_pca_nan[:, 1], marker='v', s=150, linewidths=2,
                facecolors='none', edgecolors='#9467BD',label="Hyperimpute")

sns.scatterplot(x=X_filled_missforest_pca_nan[:, 0], y=X_filled_missforest_pca_nan[:, 1], marker='^', s=150,linewidths=2,
                facecolors='none', edgecolors='#B3DE69',label="MissForest")

sns.scatterplot(x=X_filled_gain_pca_nan[:, 0], y=X_filled_gain_pca_nan[:, 1], marker='>', s=150,linewidths=2,
                facecolors='none', edgecolors='#FFC0D9',label="Gain")

sns.scatterplot(x=X_filled_sinkhorn_pca_nan[:, 0], y=X_filled_sinkhorn_pca_nan[:, 1], marker='<', s=150,linewidths=2,
                 facecolors='none', edgecolors='#5FBDFF',label="Sinkhorn")

sns.scatterplot(x=X_filled_mean_pca_nan[:, 0], y=X_filled_mean_pca_nan[:, 1], marker='s', s=150,linewidths=2,
                facecolors='none',edgecolors='#FDBF6F',label="Mean")


sns.scatterplot(x=X_filled_et_pca_nan[:, 0], y=X_filled_et_pca_nan[:, 1], marker='*', s=300, linewidths=0,
                color='#F35F5F', edgecolors='#F35F5F',label="MatImpute")

plt.legend(fontsize=18, ncol=2)
# set the xticks and yticks size
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.xlabel("PC1 ({:.2f})".format(pca_1_ratio), fontsize=24)
plt.ylabel("PC2 ({:.2f})".format(pca_2_ratio), fontsize=24)
plt.tight_layout()
plt.savefig("results/PCA-analysis-pue.png", dpi=300)
plt.show()
```
