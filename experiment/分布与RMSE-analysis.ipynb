```python
df = pd.DataFrame()

for file in files:
    df = pd.concat([df, pd.read_excel(file)], axis=0)
df
```

```python
df.dropna(inplace=True)
df
```

```python
result = df.groupby(['Method', 'Range'])['RMSE'].mean()
result
```

```python
plot_df = pd.DataFrame(columns=["Method", "Range", "RMSE"])
methods = df['Method'].unique()
for method in methods:
    for range in ["≤1σ", "1~2σ", ">2σ"]:
        plot_df = pd.concat(
            [plot_df, pd.DataFrame([[method, range, result[method][range]]], columns=["Method", "Range", "RMSE"])],
            axis=0)
plot_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

import scienceplots

plt.style.use(['nature'])
plt.rcParams["font.family"] = "Times New Roman"

warnings.filterwarnings("ignore")

my_palette = sns.color_palette(["#F35F5F", "#7ED9D9", "#9467BD", "#B3DE69", "#FFC0D9", "#5FBDFF", "#FDBF6F"])

fig = plt.figure(figsize=(8, 12))
axes = fig.subplots(3, 1, sharex=True, sharey=True)

axs = axes.flatten()

ax = axs[0]
df = pd.read_excel('std-rmse-bandgap.xlsx')
df = df[df['Col'] == 'sum_Density_(g/mL)']
plot_df = df[['Method', 'Range', 'RMSE']]
plot_df.reset_index(drop=True, inplace=True)
print(plot_df)
sns.barplot(x='Range', y='RMSE', hue='Method', data=plot_df, palette=my_palette, order=["≤1σ", "1~2σ", ">2σ"], ax=ax,saturation=1)
# add a horizontal line at y=1
ax.axhline(y=1, color='#E72929', linestyle='--')
# delete legend
ax.get_legend().remove()
# set title inside the plot
ax.text(0.5, 0.85, "sum_Density_(g/mL) in Band Gap", horizontalalignment='center', fontsize=20, transform=ax.transAxes)
# delete y-axis label
ax.set_ylabel('')
ax.tick_params(labelsize=20)
# x label size
ax.set_xlabel(ax.get_xlabel(), fontsize=24)

ax = axs[1]
df = pd.read_excel('std-rmse-bandgap.xlsx')
df.dropna(inplace=True)
result = df.groupby(['Method', 'Range'])['RMSE'].mean()
plot_df = pd.DataFrame(columns=["Method", "Range", "RMSE"])
methods = df['Method'].unique()
ranges = df['Range'].unique().tolist()
# sort ranges in order of ≤1σ, 1~2σ, >2σ
ranges = sorted(ranges, key=lambda x: 0 if x == "≤1σ" else 1 if x == "1~2σ" else 2)
for method in methods:
    for range in ranges:
        plot_df = pd.concat(
            [plot_df, pd.DataFrame([[method, range, result[method][range]]], columns=["Method", "Range", "RMSE"])],
            axis=0)
print(plot_df)
if ">2σ" not in plot_df['Range'].values:
    sns.barplot(x='Range', y='RMSE', hue='Method', data=plot_df, palette=my_palette, order=["≤1σ", "1~2σ"], ax=ax,saturation=1)
else:
    sns.barplot(x='Range', y='RMSE', hue='Method', data=plot_df, palette=my_palette, order=["≤1σ", "1~2σ", ">2σ"],
                ax=ax,saturation=1)
# add a horizontal line at y=1
ax.axhline(y=1, color='#E72929', linestyle='--')
# delete legend
ax.get_legend().remove()
# set title inside the plot
ax.text(0.5, 0.85, names[files.index(file)], horizontalalignment='center', fontsize=20, transform=ax.transAxes)
# delete y-axis label
ax.set_ylabel('')
ax.tick_params(labelsize=20)
# x label size
ax.set_xlabel(ax.get_xlabel(), fontsize=24)

ax = axs[-1]
df = pd.DataFrame()
for file in files:
    df = pd.concat([df, pd.read_excel(file)], axis=0)
df.dropna(inplace=True)
result = df.groupby(['Method', 'Range'])['RMSE'].mean()
plot_df = pd.DataFrame(columns=["Method", "Range", "RMSE"])
methods = df['Method'].unique()
for method in methods:
    for range in ["≤1σ", "1~2σ", ">2σ"]:
        plot_df = pd.concat(
            [plot_df, pd.DataFrame([[method, range, result[method][range]]], columns=["Method", "Range", "RMSE"])],
            axis=0)
sns.barplot(x='Range', y='RMSE', hue='Method', data=plot_df, palette=my_palette, order=["≤1σ", "1~2σ", ">2σ"],
            ax=ax,saturation=1)
# add a horizontal line at y=1
ax.axhline(y=1, color='#E72929', linestyle='--')

# delete legend
ax.get_legend().remove()
# set title inside the plot
ax.text(0.5, 0.85, "Average over all datasets", horizontalalignment='center', fontsize=20, transform=ax.transAxes)
# delete y-axis label
ax.set_ylabel('')
ax.tick_params(labelsize=20)
# x label size
ax.set_xlabel(ax.get_xlabel(), fontsize=24)
# set common legend bottom center
fig.text(0.02, 0.5, 'RMSE / σ', va='center', rotation='vertical', fontsize=24)
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', fontsize=20, bbox_to_anchor=(0.52, 0.05), ncol=3, frameon=False)
plt.tight_layout()
fig.subplots_adjust(hspace=0.08, wspace=0.02, left=0.10, bottom=0.21)
plt.savefig('results/std-rmse-all.png', dpi=300, bbox_inches='tight')
```
